```python
# Data manipulation
import numpy as np
import pandas as pd
np.random.seed(2)
pd.options.display.max_columns = None

# Plotting
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
%matplotlib inline
```

Below are the results for each of our three models for each metric

| Choice          | Precision       | Recall          | Accuracy        | F1 score      | ROC AUC     |
|-----------------|-----------------|-----------------|-----------------| --------------| ------------|
| RNN             | 0.6225          | 0.7250          | 0.5474          | 0.6238        | 0.6351      |
| LGB             | 0.7765          | 0.8143          | 0.7591          | 0.7950        | 0.7496      |
| XGB             | 0.0000          | 0.0000          | 0.0000          | 0.0000        | 0.0000      |

```python
metrics = ['precision', 'recall', 'accuracy', 'f1', 'roc_auc']

rnn_scores = [0.6225, 0.7250, 0.5474, 0.6238, 0.6351]
lgb_scores = [0.7765, 0.8143, 0.7591, 0.7950, 0.7496]

rnn_pd = pd.DataFrame({'metric': metrics, 'score': rnn_scores, 'model': 'RNN'})
lgb_pd = pd.DataFrame({'metric': metrics, 'score': lgb_scores, 'model': 'LGB'})

all_scores = pd.concat([rnn_pd, lgb_pd])

# Barplot of scores
f = plt.figure(figsize=(20,12))
g = sns.barplot(x = 'score', 
                y = 'metric', 
                hue = 'model', 
                data = all_scores)
g = g.set(xlabel='Score', ylabel='Metric', title='Comparison of all 3 models')
```
